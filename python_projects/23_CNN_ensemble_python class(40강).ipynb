#### ※ 객체 지향에 대한 간단한 설명 
* 객체 → 기능 + 데이터가 같이 존재하는 하나의 덩어리 
* 하나의 덩어리 : 일정크기의 메모리 공간 
* 객체 = instance = object 
* 객체를 만들고 싶어요!
> * 객체가 가지는 기능이 어떠한 기능인지, 
저장할 데이터가 어떻게 저장되는지를 
알아야 객체를 만들 수 있다.
> * 객체를 만들기 위해서는 만들 객체에 대한 정보가
(어떤 기능이 있는지, 데이터를 어떻게 저장할 것인지) 필요하다. 

* 객체를 만들기 위해서 필요한 정보를 기술해 놓은 것 
  → Class
* 객체를 생성하기 위해서 class부터 정의해보자



```
## class definition
class Car:
    #생성자 : class로부터 객체가 만들어질때 무조건 호출되는 함수 
    def __init__(self,name): #self : 현재 사용되는 객체에 대한 reference(생성자에 기본적으로 들어간다)
        self.myName = name #self가 reference하는 곳에 myName이란 곳을 하나 만들음. 거기에 name 할당 
        self.myList = [1,2,3,4] #하나의 객체가 여러가지 공간을 갖게 됨 
                                #객체를 '추상 데이터 타입' 이라고 함.(여러 데이터타입을 같이 가질 수 있음)
                                #→프로그래밍 할 때 엄청난 유연성을 발휘할 수 있다. 
        print("{} 객체가 생성되었어요!!".format(self.myName) )
        print("{} 얘도 출력되나".format(self.myList))
car1 = Car("소나타") #소나타라는 글자가 name으로 전달된다 (name = "소나타")
                     #객체를 생성하고 변수에 저장
car2 = Car("그랜저")
 
```
