```
import urllib.request 
res = urllib.request.urlopen("http://www.naver.com/")
print(type(res)) # http.client.HTTPResponse 클래스 
print(res.status)
print(res.version) #HTTP protocol version used by server: 10 for HTTP/1.0,  11 for HTTP/1.1
print(res.msg)
res_header = res.getheaders()
print("[header 정보]##############")
for s in res_header:
    print(s)
```

```
import urllib.request
import urllib.parse
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
# 매개변수를 URL 인코딩합니다. 
values = {
    'stnId': '108'   
} #stnId는 기상 정보를 알고 싶은 지역
params = urllib.parse.urlencode(values)
# 요청 전용 URL을 생성합니다.  
url = API + "?" + params
print("url=", url)
# 다운로드합니다.  
data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```

```
import urllib.request 
res = urllib.request.urlopen("http://70.12.116.160:8080/login/login.html") #로컬 도메인은 다른 IDE 사용해서 확인할 것 
print(res)
res_header = res.getheaders()
print("[header 정보]##############")
for s in res_header:
    print(s)

print("[body 정보]##############")
print(res.read().decode('utf-8')) 
```

```
import urllib.request
url = 'http://www.python.org/'
f = urllib.request.urlopen(url)
print(type(f))
print(type(f.info()))
encoding = f.info().get_content_charset()
print(url, '페이지 인코딩 정보:', encoding)
text = f.read(500).decode(encoding)
print(text)
```

```
#url주소 해석 parse모듈 실습 
from urllib.parse import urlparse 
url2 = urlparse('https://docs.python.org/ko/3/library/urllib.request.html#module-urllib.request')
print('도메인 정보 : ', url2.netloc)
print('패스 정보  : ', url2.path)
print('쿼리 문자열 정보  : ', url2.query)
print('uri 스킴 정보  : ', url2.scheme)
print('포트 정보 : ', url2.port)
print('프레그먼트 정보 : ', url2.fragment)
print('URL 문자열 정보 : ', url2.geturl())
print('URL 객체 정보 : ', url2)

#상기된 url을 보고 그대로 받아온것 (안적힌 내용들은 당연히 안적혀있으니깐))
```
